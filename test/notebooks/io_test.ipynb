```python
df = pd.read_csv('../../data/geoid.gdf', skiprows=36, header=None, sep=' \t')
df
```

```python
re.split('\s+', df.iloc[0, 0])
```

```python
def filter(string):
    split_kwrg = re.split('\s+', string[0])
    return [float(split_kwrg[0]),
                       float(split_kwrg[1]),
                       float(split_kwrg[2])]
```

```python
new_df = pd.DataFrame(columns=['long', 'lat', 'geoid'])
```

```python
new_df = df.apply(filter, axis=1)
```
